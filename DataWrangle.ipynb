# Selecting Cells

```python
whr23.iloc[:,0:2]
```

```python
whr23.iloc[2,1]
```

```python
whr23.loc[20]
```

```python
whr23.loc[[20], ["country_name", "ladder_score", "continent"]]
```

```python
whr23.loc[whr23["continent"] == "Europe"]
```

```python
whr23.loc[whr23["ladder_score"] > 7]
```

# Bivariate Relationships

```python
g = sns.PairGrid(whr23, diag_sharey = False)
g.map_upper(sns.scatterplot, s = 15)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw = 2)
```

```python
whr23.corr(numeric_only = True)
```

```python
sns.heatmap(whr23.corr(numeric_only = True), annot = True)
plt.rcParams["figure.figsize"] = (40,7)
plt.show()
```
